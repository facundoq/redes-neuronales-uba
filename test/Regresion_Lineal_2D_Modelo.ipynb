# Visualización del modelo y la superficie del error

El siguiente código carga un conjunto de datos, y luego crea un modelo de regresión lineal para predecir los valores de salida `y` en base a los de entrada `x1` y `x2` y en base a la función `y=m1 x1+ m2 x2+b`.

Probá modificando los parámetros `m1` y `m2` y `b` del modelo, tratando de minimizar *a mano* el error. Observá como cambia la recta `m1 x1 + m2 x2+b` y qué tan bien se ajusta a los datos.  

En este caso, ya no podemos visualizar la función de error ya que hay 3 parámetros y por ende necesitaríamos 4 dimensiones (3 para los parámetros y una para el valor del error en base a ellos).



```python
import numpy as np
## Carga de datos
data = rnutil.load_dataset_numpy("study_regression_2d_small.csv")

x1,x2,y=data[:,0],data[:,1],data[:,2]

## Parámetros del modelo (probar valores entre -5 y 5)
m1=3
m2=-2
b=2

#Calculo del error del modelo
y_predicted= x1*m1+x2*m2+b
errors= (y-y_predicted)**2
mean_error=errors.mean()

rnutil.plot_regresion_lineal(np.array([m1,m2]),b,x1,x2,y,"x1 (Horas estudiadas)","x2 (Promedio)","y (Nota)",title="Modelo Final")
import matplotlib.pyplot as plt
plt.show()
```
